#**Gráficos utilizando la librería Matplotlib**

```
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
```

## **Guardar la figura**

La función savefig permite guardar la fígura en un formato de imagen que por lo general es png

```
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x**2)
plt.savefig("my_square_function.png", transparent=True)
```

## **Dibujar varios graficos en una misma figura (Subplots)**

* Para crear una figura con multiples graficos se utiliza la funcion **subplot**, la cual permite organizar dichos graficos en una cuadricula (grid).

* La funcion subplot recibe como parametros la cantidad de filas y columnas de la figura y el indice de la grafica que se desea dibujar en dicha posicion.

* Las posiciones se indican de izquierda a derecha y de arriba a abajo.

```
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot = bottow left
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot = bottom right
plt.plot(x, x**4)
plt.show()
```

Algunos graficos necesitan extenderse por mas de una posicion en la cuadricula 

```
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 1, 2)  # 2 rows, *1* column, 2nd subplot = bottom
plt.plot(x, x**3)
plt.show()
```

para posicionar graficos dentro de una misma figura de una forma mas compleja, se puede utilizar la funcion **subplot2gird**. En este caso se deben especificar el numero de filas y columnas en la cuadricula, la position de la grafica dentro de la cuadricula, y opcionalmente cuantas filas y columnas ocupara dicha grafica.





```
plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3,3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3,3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

## **Multiples Figuras**

la funcion **figure** permite crear diferentes figuras que pueden contener diferentes graficos. Es es util cuando se quieren guardar graficos en diferentes figuras.

```
x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Square and Cube")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")

plt.figure(1)      # back to figure 1, current subplot is 212 (bottom)
plt.plot(x, -x**3, "r:")

plt.show()
```

# **Insertar texto en los graficos**

```
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='blue', horizontalalignment="center")
plt.text(px - 0.08, py, "Beautiful point", ha="right", weight="heavy")
plt.text(px, py, "x = %0.2f\ny = %0.2f"%(px, py), rotation=50, color='gray')

plt.show()
```

La funcion ***annotate*** permite insertar anotaciones dentro de un grafico

```
plt.plot(x, x**2, px, py, "ro")
plt.annotate("Beautiful point", xy=(px, py), xytext=(px-1.3,py+0.5),
                           color="green", weight="heavy", fontsize=14,
                           arrowprops={"facecolor": "lightgreen"})
plt.show()
```

Es posible agregar mas efectos o formas al texto dentro de un grafico

```
plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "Beautiful point", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='black', ha="center", bbox=bbox_props)

plt.show()
```

# **leyendas (legends)**

para insertar leyendas en los graficos, primero se deben definir las etiquetas (labels)

```
x = np.linspace(-1.4, 1.4, 50)
plt.plot(x, x**2, "r--", label="Square function")
plt.plot(x, x**3, "g-", label="Cube function")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

# **Escala de los ejes**

Matplotlib permite escala no lineal en los ejes de los graficos, como por ejemplo la escala logaritmica

```
x = np.linspace(0.1, 15, 500)
y = x**3/np.exp(2*x)

plt.figure(1)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

plt.figure(2)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

plt.figure(3)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.figure(4)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.05)
plt.title('symlog')
plt.grid(True)

plt.show()
```

# **Configuraciones sobre los ejes, lineas y los labels**

Los ejes tienen una serie de atributos llamados "ticks". Existen otros atributos llamados "tickers" que se encargan de decidir donde posicionar los ticks

```
x = np.linspace(-2, 2, 100)

plt.figure(1, figsize=(15,10))
plt.subplot(131)
plt.plot(x, x**3)
plt.grid(True)
plt.title("Default ticks")

ax = plt.subplot(132)
plt.plot(x, x**3)
ax.xaxis.set_ticks(np.arange(-2, 2, 1))
plt.grid(True)
plt.title("Manual ticks on the x-axis")

ax = plt.subplot(133)
plt.plot(x, x**3)
plt.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom='off')
ax.xaxis.set_ticks([-2, 0, 1, 2])
ax.yaxis.set_ticks(np.arange(-5, 5, 1))
ax.yaxis.set_ticklabels(["min", -4, -3, -2, -1, 0, 1, 2, 3, "max"])
plt.title("Manual ticks and tick labels\n(plus minor ticks) on the y-axis")

plt.grid(True)

plt.show()
```

# **Graficos en 3D**

Para realizar graficos en 3D se necesita importar **Axes3D**. Luego se debe crear un subplot configurando el atributo "projection" como 3d. Esto returna un objeto **Axes3DSubplot** el cual puede ser usado llamadno la funcion **plot_surface**

```
import matplotlib
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z =(X**2 - Y**2)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

Otra forma de visualizar una gradica en 3D es por medio de la funcion **contour**

```
plt.contourf(X, Y, Z, cmap=matplotlib.cm.coolwarm)
plt.colorbar()
plt.show()
```

# **Graficar la relacion entre dos conjuntos de datos (Scatter Plots)**

```
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

Es posible modificar la apariencia de cada punto dentro del grafico

```
x, y, scale = rand(3, 100)
scale = 500 * scale ** 5
plt.scatter(x, y, s=scale)
plt.show()
```

# **Dibujar Lineas sobre los graficos**

```
from numpy.random import randn

def plot_line(axis, slope, intercept, **kargs):
    xmin, xmax = axis.get_xlim()
    plt.plot([xmin, xmax], [xmin*slope+intercept, xmax*slope+intercept], **kargs)

x = randn(1000)
y = 0.5*x + 5 + randn(1000)*2
plt.axis([-2.5, 2.5, -5, 15])
plt.scatter(x, y, alpha=0.2)
plt.plot(1, 0, "ro")
plt.vlines(1, -5, 0, color="red")
plt.hlines(0, -2.5, 1, color="red")
plot_line(axis=plt.gca(), slope=0.5, intercept=5, color="magenta")
plt.grid(True)
plt.show()
```

# **Histogramas**

Un histograma es un grafico que permite visualizar un conjunto de datos mediante barras de diferentes alturas. Cada barra es definida por un rango de valores (ancho) y la altura. Las barras de mayor altura indican que la mayoria de los datos se encuentran en su rango.

```
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211)
plt.hist(data, bins = 10, rwidth=0.8)

plt.subplot(212)
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth=0.95)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

# **Graficar Imagenes**

Para leer una imagen se debe importar **matplotlib.image** y utilizar la fun cion **imread** con el nombre del archivo como parametro. Esta funcion returna los datos de la imagen en formato de **NumPy array**.


```
import matplotlib.image as mpimg

img = mpimg.imread('Perro_1.jpg')
print(img.shape, img.dtype)
```

La imagen cargada tiene 288x432 pixels. Cada pixel esta representado por 4 elementos: rojo, verde, azul, nivel de alpha. Todo se encuentra almacenado en formato float de 32 bits con valores entre 0 y 1.

Para visualizar la imagen se utiliza la funcion **imshow**

```
plt.imshow(img)
plt.show()
```

Para ocultar los ejes de la figura:

```
plt.imshow(img)
plt.axis('off')
plt.show()
```

```
img2 = img[200:500,200:500]
print(img2.shape, img2.dtype)
```

```
plt.imshow(img2)
plt.axis('off')
plt.show()
```

De esta manera es facil generar una figura propia

```
img = np.arange(100*100).reshape(100, 100)
print(img)
plt.imshow(img)
plt.show()
```

```
plt.imshow(img, cmap="hot")
plt.show()
```

Es posible generar una imagen directamente

```
img = np.empty((20,30,3))
img[:, :10] = [1, 0, 0]
img[:, 10:20] = [0, 1, 0]
img[:, 20:] = [0, 0, 1]
plt.imshow(img)
plt.show()
```
